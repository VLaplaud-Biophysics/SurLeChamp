Le but de ce notebook est de résoudre différente versions des équations de prédation de Lotka-Voltera, appliquée a l'évolution de la population de deux armées lors d'une bataille.

```python
import numpy as np 
import scipy.integrate as scpint
import matplotlib.pyplot as plt
```

### Fonctions utiles

```python
""" Test de positivité strict"""
def strctpos(x):
    
    if x > 0:
        p = 1
    else:
        p = 0 
    
    return(p)     
```

## Résolution des équations

### Lotka-Voltera

#### Définition et résolution

```python
def Solve_LV(t,X0,P):
    
    """ Equations d'évolution """
    a,b,c,d = P
    def dX_dt_LV(X, t=0):

        du_dt =  a*X[0] -   b*X[0]*X[1] # du/dt = u(t)*a - u(t)*v(t)*b
        dv_dt =  -c*X[1] + d*X[0]*X[1] # dv/dt = -v(t)*c + v(t)*u(t)*d

        return np.array([ du_dt , dv_dt ])
    
    """ Intégration """
    X_LV, infodict = scpint.odeint(dX_dt_LV, X0, t, full_output=True) 
    
    return(X_LV)
  
                  
```

#### Graphe des solutions

```python
def PlotRes_LV(t,X,P,labels):

    u, v = X.T # Récupération des trajectoires

    print('Paramètres : \nReproduction proie = ' + str(P[0]) + '\nMortalité proie par prédateur = ' + str(P[1]) 
    + '\nMortalité prédateur = ' + str(P[2]) + '\nReproduction prédateur par proie = ' + str(P[3])) 

    """ Graphe d'évolution en temps """
    f1, ax1 = plt.subplots(dpi=250,facecolor='w')
    plt.title('Evolution des populations de proies et prédateurs')
    ax1.plot(t, u, 'r-', label=labels[0])
    ax1.plot(t, v  , 'b-', label=labels[1])
    plt.grid()
    plt.legend(loc='best')
    plt.xlabel('Temps')
    plt.ylabel('Population')
    plt.show()

    return
```

### Lanchester

#### Définition et résolution

```python
def Solve_Lc(t,X0,P):
    
    """ Equations d'évolution """
    a,b = P
    def dX_dt_Lc(X, t=0):

        du_dt =  -b*max([X[1],0])*strctpos(X[0]) # du/dt = -a*v(t) et u(t) > 0
        dv_dt =  -a*max([X[0],0])*strctpos(X[1]) # dv/dt = -b*u(t) et v(t) > 0

        return np.array([ du_dt , dv_dt ])
    
    """ Intégration """
    X_Lc, infodict = scpint.odeint(dX_dt_Lc, X0, t, full_output=True)  
    
    return(X_Lc)
```

#### Graphe des solution

```python
def PlotRes_Lc(t,X,P,labels):

    u, v = X.T # Récupération des trajectoires

    print('Paramètres : \nArmée 1 : ' + str(X[0][0]) + ' soldat, ' + str(P[0]) + ' puissance. ' + 
          '\nArmée 2 : ' + str(X[0][1]) + ' soldat, ' + str(P[1]) + ' puissance.') 

    """ Graphe d'évolution en temps """
    f1, ax1 = plt.subplots(dpi=250,facecolor='w')
    plt.title('Evolution des deux armées')
    ax1.plot(t, u, 'r-', label=labels[0])
    ax1.plot(t, v  , 'b-', label=labels[1])
    plt.grid()
    plt.legend(loc='best')
    plt.xlabel('Temps')
    plt.ylabel('Population')
    plt.show()

    return
```

### Lanchester avec reserve

```python
def Res(t):
    
    if (t>=0) & (t<2):
        R = 0
    elif (t>=2) & (t<3):
        R = 6000
    elif (t>=3) & (t<5):
        R = 0
    elif (t>=5) & (t<6):
        R = 13000
    else:
        R = 0
        
    return(R)

def Solve_LcR(t,X0,P):
    
    """ Equations d'évolution """
    a,b = P
    def dX_dt_LcR(X, t):

        du_dt =  Res(t)-b*max([X[1],0])*strctpos(X[0]) # du/dt = -a*v(t) et u(t) > 0
        dv_dt =  -a*max([X[0],0])*strctpos(X[1]) # dv/dt = -b*u(t) et v(t) > 0

        return np.array([ du_dt , dv_dt ])
   # T_LcR,X_LcR = scpint.solve_ivp(dX_dt_LcR, (0,36), X0, method='RK45', t_eval=np.linspace(0, 36,  37))
    """ Intégration """
    X_LcR, infodict = scpint.odeint(dX_dt_LcR, X0, t, full_output=True)  
    
    return(X_LcR)
```

## Résultats

### Lotka-Voltera

```python
t = np.linspace(0, 15,  1000) # Temps

""" Lotka-Voltera """
P_LV = [1, 0.1, 1.5, 0.075] # Set de paramètres 
X0_LV = np.array([10, 5])  # Conditions initiales 


print('\033[1m' + '\033[4m' + 'Lotka-Voltera :\n' + '\033[0m')

PlotRes_LV(t,Solve_LV(t,X0_LV,P_LV),P_LV,['Proie','Prédateur'])
```

### Lanchester

```python
t = np.linspace(0, 20,  1000) # Temps

""" Lanchester """
P_Lc = [0.4, 0.1] # Set de paramètres 
X0_Lc = np.array([500, 1000])  # Conditions initiales

print('\033[1m' + '\033[4m' + 'Lanchester :\n' + '\033[0m')

PlotRes_Lc(t,Solve_LcR(t,X0_Lc,P_Lc),P_Lc,['Armée 1','Armée 2'])
```
